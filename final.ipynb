**T-test for independent samples**


𝐻_0:µ_1=µ_2  ("there is no significant difference in median value of houses bounded by the Charles river.")
𝐻_1:µ_1≠µ_2 ("there is a significant difference in median value of houses bounded by the Charles river.")
α = 0.05

```python
import scipy.stats
scipy.stats.ttest_ind(boston_df[boston_df['CHAS'] == 0.0]['MEDV'],
                   boston_df[boston_df['CHAS'] == 1.0]['MEDV'],equal_var= False)
```

Since the p-value is less than alpha value 0.05, we reject the null hypothesis as there is enough proof that there is a statistical difference in in median value of houses bounded by the Charles river.

############################################################

**ANOVA**

*   $H\_0: µ\_1 = µ\_2 = µ\_3$ (the three population means are equal)
*   $H\_1:$ At least one of the means differ

```python
under35_age = boston_df[boston_df['age'] == 'under35']['MEDV']
for35to70_age = boston_df[boston_df['age'] == '35to70']['MEDV']
over70_age= boston_df[boston_df['age'] == 'over70']['MEDV']
```

```python
f_statistic, p_value = scipy.stats.f_oneway(under35_age, for35to70_age, over70_age)
print("F_Statistic: {0}, P-Value: {1}".format(f_statistic,p_value))
```

Since the p-value is less than 0.05, we will reject the null hypothesis as there is significant evidence that at least one of the means differ.
so the there is a significant difference in MEDV and AGE.

**Pearson Correlation**

State the hypothesis:
*   $H\_0:$ Nitric oxide concentrations is not correlated with proportion of non-retail business acres per town
*   $H\_1:$ Nitric oxide concentrations is correlated with proportion of non-retail business acres per town

```python
ax = sns.scatterplot(x="NOX", y="INDUS", data=boston_df)
```

```python
scipy.stats.pearsonr(boston_df['NOX'], boston_df['INDUS'])
```

Since the p-value  < 0.05, we reject  the Null hypothesis and conclude that there  exists a relationship between  Nitric oxide concentrations and proportion of non-retail business acres per town.
